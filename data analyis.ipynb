```python
import matplotlib.pyplot as plt
data=pd.read_csv('tips.csv')
plt.scatter(data['day'],data['tip'])
plt.title('Scatter plot: Tips per day visualization')
plt.ylabel('tips')
plt.xlabel('day')
plt.show()
```

```python
import matplotlib.pyplot as plt
data=pd.read_csv('tips.csv')
plt.scatter(data['day'],data['tip'], c=data['size'], s=data['total_bill'])
plt.title('Scatter plot: Tips per day visualization')
plt.ylabel('tips')
plt.xlabel('day')
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = x*2

# first plot with X and Y data
plt.plot(x, y)

x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]

# second plot with x1 and y1 data
plt.plot(x1, y1, '-.')

plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title('multiple plots')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = x*2

plt.plot(x, y)

x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]

plt.plot(x, y1, '-.')
plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title('multiple plots')

plt.fill_between(x, y, y1, color='green', alpha=0.5)
plt.show()
```
