A stem plot, also known as a stem-and-leaf plot, is a type of plot used to display data along a number line. Stem plots are particularly useful for visualizing discrete data sets, where the values are represented as stems extending from a baseline, with data points indicated as leaves along the stems. lets understand the components of a typical stem plot:

Stems: The stems represent the main values of the data and are typically drawn vertically along the y-axis.
Leaves: The leaves correspond to the individual data points and are plotted horizontally along the stems.
Baseline: The baseline serves as the reference line along which the stems are drawn.

```python
# importing libraries 
import matplotlib.pyplot as plt 
import numpy as np 

x = np.linspace(0.1, 2 * np.pi, 41) 
y = np.exp(np.sin(x)) 

plt.stem(x, y) 
plt.show() 
```

Stackplot, also known as a stacked area plot, is a type of plot that displays the contribution of different categories or components to the total value over a continuous range, typically time. It is particularly useful for illustrating changes in the distribution of data across multiple categories or groups.

```python
import matplotlib.pyplot as plt

# List of Days
days = [1, 2, 3, 4, 5]

# No of Study Hours
Studying = [7, 8, 6, 11, 7]

# No of Playing Hours
playing = [8, 5, 7, 8, 13]

# Stackplot with X, Y, colors value
plt.stackplot(days, Studying, playing,
            colors =['r', 'c'])

# Days
plt.xlabel('Days')

# No of hours
plt.ylabel('No of Hours')

# Title of Graph
plt.title('Representation of Study and \
Playing wrt to Days')

# Displaying Graph
plt.show()
```

A box plot, also known as a box-and-whisker plot, provides a visual summary of the distribution of a dataset. It represents key statistical measures such as the median, quartiles, and potential outliers in a concise and intuitive manner. Box plots are particularly useful for comparing distributions across different groups or identifying anomalies in the data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create a noisy dataset
np.random.seed(0)
data = np.random.normal(loc=50, scale=10, size=1000)  # 1000 data points around mean=50 with std=10
noise = np.random.normal(loc=0, scale=5, size=1000)  # Add noise
noisy_data = data + noise

# Step 2: Apply binning to reduce noise
# Using equal-width binning to create 10 bins
bins = np.linspace(min(noisy_data), max(noisy_data), 11)  # 10 bins
binned_data = pd.cut(noisy_data, bins, include_lowest=True)

# Step 3: Plot the original noisy data and the binned data
plt.figure(figsize=(12, 6))

# Plot original noisy data
plt.subplot(1, 2, 1)
plt.hist(noisy_data, bins=30, color='gray', alpha=0.7)
plt.title("Original Noisy Data")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Plot binned data
plt.subplot(1, 2, 2)
binned_counts = pd.value_counts(binned_data).sort_index()
binned_counts.plot(kind='bar', color='skyblue')
plt.title("Binned Data (Equal-width Binning)")
plt.xlabel("Binned Intervals")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Example time series data (12 months)
data = pd.Series([120, 135, 148, 156, 165, 189, 210, 250, 285, 310, 350, 400], 
                 index=pd.date_range('2022-01-01', periods=12, freq='M'))

# Apply seasonal decomposition with quarterly seasonality (period=4)
decomposition = sm.tsa.seasonal_decompose(data, model='additive', period=4)

# Plot the decomposition
decomposition.plot()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample time series data
data = pd.Series([3, 5, 2, 8, 7, 10], index=pd.date_range('2024-01-01', periods=6, freq='M'))

# Calculating rolling mean
rolling_mean = data.rolling(window=3).mean()

plt.plot(data, label='Original Data')
plt.plot(rolling_mean, label='Rolling Mean', color='orange')
plt.legend()
plt.show()
```
